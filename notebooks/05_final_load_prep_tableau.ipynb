## 05: Final Load Prep & KPI Validation

**Objective:** To perform a final quality audit on the enriched dataset and export the definitive version for Tableau. We will also compute the "Ground Truth" KPIs in Python to ensure our Tableau dashboard metrics are accurate.

### Environment Setup

```python
import pandas as pd
import numpy as np

# Load the cleaned dataset from the previous phase
df = pd.read_csv('/content/drive/MyDrive/DVA_2/EV_Population_Cleaned.csv')

print(f"Final dataset loaded with {df.shape[0]} rows and {df.shape[1]} columns.")
```

### Data Integrity Audit for Tableau
Before uploading to Tableau, we must ensure that columns used for Maps and Filters are perfect. Tableau can crash or show "Null" clusters if coordinates are missing.

```python
# 1. Verification of Geospatial Data
print("--- Geospatial Integrity Check ---")
null_coords = df[['Latitude', 'Longitude']].isnull().sum()
print(f"Missing Coordinates:\n{null_coords}")

# 2. Verification of Data Types
# Ensure Postal Code and Legislative District are strings (Tableau handles them better as Dimensions)
df['Postal Code'] = df['Postal Code'].astype(str)
df['Legislative District'] = df['Legislative District'].astype(str).replace('nan', 'Unknown')

# 3. Check for Infinite Values (which can happen in some calculations)
print(f"\nAny Infinite Values: {np.isinf(df.select_dtypes(include=np.number)).values.any()}")
```

### KPI Checklist (The "Ground Truth")
Calculate your primary KPIs here. When you build your Tableau dashboard, these numbers must match. If they don't, you know your Tableau filters are wrong.

```python
print("--- PROJECT KPI SUMMARY (Ground Truth) ---")

# KPI 1: Total EV Population
total_ev = len(df)
print(f"Total EV Population: {total_ev:,}")

# KPI 2: Market Share %
luxury_share = (df[df['Market_Segment'] == 'Luxury'].shape[0] / total_ev) * 100
print(f"Luxury Segment Market Share: {luxury_share:.2f}%")

# KPI 3: Average Range (Verified Only)
avg_range = df[df['Electric Range Raw'] > 0]['Electric Range Raw'].mean()
print(f"Average Electric Range (Verified): {avg_range:.2f} miles")

# KPI 4: Infrastructure Complexity
high_complexity_zones = len(df[df['Utility_Complexity_Score'] > 1])
complexity_perc = (high_complexity_zones / total_ev) * 100
print(f"Vehicles in Multi-Utility Zones: {complexity_perc:.2f}%")
```
